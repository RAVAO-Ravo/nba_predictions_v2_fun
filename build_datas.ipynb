# Importation des modules

```python
# Importation des modules
from MakeDataset import make_dataset
from FeatureEngineering import create_features_and_labels
from PlotEval import plot_1d, plot_2d
from CustomSplitting import train_test_split_twin

# Importation des variables globales
from config import (
    test_size,
    random_state,
    data_file,
    x_train_file,
    x_test_file,
    y_train_file,
    y_test_file
)
```

# Création des données

```python
# Importation de la fonction make_dataset
data = make_dataset()

# Enregistrement des données dans un fichier CSV
data.to_csv(path_or_buf=data_file, index=False)

# Affichage d'un échantillon aléatoire de 10 lignes des données
display(data.sample(n=10, random_state=random_state))
```

# Visualisation des données

```python
# Création des caractéristiques (features) et des étiquettes (target) à partir des données
features, target = create_features_and_labels(data=data)

# Affichage d'un échantillon aléatoire de 10 lignes des caractéristiques (features)
display(features.sample(n=10, random_state=random_state))

# Tracé d'un graphique 1D pour visualiser la relation entre les caractéristiques et la cible
plot_1d(features=features, target=target)

# Tracé d'un graphique 2D pour visualiser la relation entre deux caractéristiques et la cible
plot_2d(X=features, y=target)
```

# Splitting des données

```python
# Division des données en ensembles d'entraînement et de test
X, x_test, y, y_test = train_test_split_twin(features=features, target=target, test_size=test_size, random_state=random_state)

# Tracé d'un graphique 1D
plot_1d(X, y)

# Tracé d'un graphique 1D
plot_1d(x_test, y_test)

# Tracé d'un graphique 2D
plot_2d(X, y)

# Tracé d'un graphique 2D
plot_2d(x_test, y_test)

# Enregistrement des caractéristiques d'entraînement dans un fichier CSV
X.to_csv(x_train_file, index=False)

# Enregistrement de la cible d'entraînement dans un fichier CSV
y.to_csv(y_train_file, index=False)

# Enregistrement des caractéristiques de test dans un fichier CSV
x_test.to_csv(x_test_file, index=False)

# Enregistrement de la cible de test dans un fichier CSV
y_test.to_csv(y_test_file, index=False)
```
